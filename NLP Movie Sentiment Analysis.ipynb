```python
#Importing the libraries
import nltk
import pandas as pd
import sklearn
import random
```

```python
#Loading the dataset
movies=pd.read_csv('D:IMDB/datasetm.csv')
print movies
```

```python
#removing punctation and making all capital letter lowercase
mst=movies['SentimentText']
msp=mst.str.replace(pat=r'([.;:!\'?,\"*()\[\]])',repl=' ')
mspt=msp.str.replace(pat='(<br\s*/><br\s*/>)|(\-)|(\/)',repl=' ')
msptl=mspt.str.lower()
print msptl
```

```python
#Creating the test and train set for the model later
from sklearn.utils import shuffle
msptl=shuffle(msptl)
tests=int(len(msptl)*0.7)
trains=int(len(msptl)*0.3)
test =msptl.iloc[:tests]
train = msptl.iloc[:trains]
print (train)
```

```python
#tokenizing each word for the stemming and lemmization
mtok=msptl.apply(nltk.word_tokenize)
```

```python
#Removing stop words using custom list from the tokenized list
stop = ['in','of','at','a']
mtoks=mtok.replace(to_replace=stop, value=" ",regex=True)
print mtoks
```

```python
#Stemming the tokenised filtered list
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
stemok=[]
for i in range(len(mtoks)):
    try:
        stemok.append([stemmer.stem(x) for x in mtoks[i]])
    except:
        pass
print stemok
```

```python
#lemmatizing the tokenised filtered list
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemok=[]
for i in range(len(mtoks)):
    try:
        lemok.append([lemmatizer.lemmatize(y) for y in mtoks[i]])
    except:
        pass
print lemok
```
