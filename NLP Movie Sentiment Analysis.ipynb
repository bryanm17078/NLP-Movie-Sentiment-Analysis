```python
#Importing the libraries
import nltk
import pandas as pd
import sklearn
import random
import numpy as np
```

```python
#Loading the dataset
movies=pd.read_csv('D:IMDB/datasetm.csv')
print movies
```

```python
#removing punctation and making all capital letter lowercase
mst=movies['SentimentText']
msp=movies['SentimentText'].str.replace(pat=r'([.;:!\'?,\"*()\[\]])',repl=' ')
mspt=movies['SentimentText'].str.replace(pat='(<br\s*/><br\s*/>)|(\-)|(\/)',repl=' ')
msptl=movies['SentimentText'].str.lower()
cmovies=movies
cmovies['CSentimentText']=msptl
cmovies=cmovies.drop(columns='SentimentText')
print cmovies
```

```python
#Creating the test and train set for the model later
from sklearn.utils import shuffle
msptl=shuffle(msptl)
tests=int(len(msptl)/2)
trains=int(len(msptl)/2)
test =msptl.iloc[:tests]
train = msptl.iloc[:trains]
print (train)
```

```python
#tokenizing each word for the stemming and lemmization
import sys
reload(sys)
sys.setdefaultencoding('utf-8')
mtok=cmovies['CSentimentText'].apply(nltk.word_tokenize)
```

```python
#Removing stop words using custom list from the tokenized list
stop = ['in','of','at','a']
mtoks=mtok.replace(to_replace=stop, value=" ",regex=True)
cmovies['TCST']=mtoks
print cmovies
print display(len(mtoks))
```

```python
#Stemming the tokenised filtered list
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
stemok=[]
for i in range(len(mtoks)):
    try:
        stemok.append([stemmer.stem(x) for x in mtoks[i]])
    except:
        stemok.append('No_stem')
#print stemok[0:100]
cmovies['StemText']=cmovies['TCST']
cmovies['StemText']=stemok
print display(len(stemok))
```

```python
#lemmatizing the tokenised filtered list
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemok=[]
for i in range(len(mtoks)):
    try:
        lemok.append([lemmatizer.lemmatize(y) for y in mtoks[i]])
    except:
        lemok.append('No_Lem')

        
cmovies['LemText']=cmovies['TCST']
cmovies['LemText']=lemok
print len(lemok)
```

```python
#Randomising the data to reduce chance of overfitting
from sklearn.utils import shuffle
cmovie=shuffle(cmovies)
tests=int(len(cmovie)/2)
trains=int(len(cmovie)/2)
test =cmovies.iloc[:tests]
train = cmovies.iloc[:trains]

train=train.sort_values(by='Sentiment')
print display((train))
```

```python
#Normal tokenised text
trains=train.drop(columns=['CSentimentText','Sentiment','StemText','LemText'])
trainl=train.drop(columns=['CSentimentText','TCST','StemText','LemText'])
#trainl=trainl.astype('int')
print display(trains)
```

```python
#Encoding the words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
cv.fit(train['CSentimentText'])
X = cv.transform(train['CSentimentText']).toarray()
print display(X)
```

```python
#Splitting the model into a training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,trainl, test_size=0.3)
print display(X_train.shape, y_train.shape)
```

```python
#Applying the model (svm)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

#using n grams to increase predictive power
ngram_vectorizer = CountVectorizer(binary=True, ngram_range=(1, 2))
ngram_vectorizer.fit(train['CSentimentText'])
X = ngram_vectorizer.transform(train['CSentimentText'])
X_test = ngram_vectorizer.transform(train['CSentimentText'])

X_train, X_val, y_train, y_val = train_test_split(
    X, trainl, train_size = 0.75
)

#Using regularisation to reduce overfitting
for c in [0.01, 0.05, 0.25, 0.5, 1]:
    #Fitting the model
    svm = LinearSVC(C=c)
    svm.fit(X_train, y_train)
    print display("Accuracy for C=%s: %s" 
           % (c, accuracy_score(y_val, svm.predict(X_val))))
    
#'Accuracy for C=0.01: 0.88064'
#'Accuracy for C=0.05: 0.88128'
#'Accuracy for C=0.25: 0.87936'
#'Accuracy for C=0.5: 0.87968'
#'Accuracy for C=1: 0.87968'

#C=0.05 has the highest accuracy
#Fitting the final model 
final_svm_ngram = LinearSVC(C=0.01)
final_svm_ngram.fit(X, trainl)
print display ("Final Accuracy: %s" 
       % accuracy_score(trainl, final_svm_ngram.predict(X_test)))



#'Final Accuracy: 0.99984'
```

```python
#Predicting sentiment
sdic=['bad','good']
mov=[]
predict=final_svm_ngram.predict(X_test)
for x in np.nditer(predict):
    mov.append(sdic[x])

    
print display(mov[0])
print display(train['CSentimentText'].iloc[0])

print display(mov[10000])
print display(train['CSentimentText'].iloc[10000])
```
